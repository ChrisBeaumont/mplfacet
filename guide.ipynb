# Faceted plots with matplotlib

[mplfacet](http://www.github.com/ChrisBeaumont/mplfacet) provides an interface to easily create faceted plots with matplotlib. Faceted plots are small multiples of plots where data are partitioned into groups based on one or more "key" values. 

For example, let's look at the mtcars dataset, which lists properties like mpg, cylinder count, horsepower, weight, etc. for a few cars

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn import datasets

from mplfacet import Facet
```

```python
with open('mtcars.csv') as infile:
    data = np.recfromcsv(infile)

print "Columns"
print "-------"
print "\n".join(data.dtype.names)
```

Let's make scatter plots of weight vs mpg, for cars grouped by their cylinder count. This creates 3 plots -- one each for the cars with 4, 6, and 8 cylinders.

```python
facet_key = data.cyl
plot_data = [data.wt, data.mpg]
f = Facet(facet_key, plot_data, xlabel='Weight', ylabel='MPG')
f.scatter()
```

Any plot method that takes two data arguments will work with this facet object. So will extra arguments and keywords:

```python
f.plot('r^', alpha=.6, ms=15)
```

To make faceted histograms, use a single data item:

```python
Facet(data.cyl, [data.mpg], xlabel='MPG').hist(bins = np.linspace(10, 35, 10))
```

To make variable-sized scatter plots, pass 3 data items

```python
f = Facet(data.cyl, [data.wt, data.mpg, data.hp], 
          xlabel='Weight', ylabel='MPG')
f.scatter(alpha=0.3)
```

The keyword arguments for `plt.subplots` can be used to change the layout. For example, to arrange vertically:

```python
f = Facet(data.cyl, [data.mpg], ncols=1, xlabel='MPG',
          figsize=(4, 6))
f.hist(bins = np.linspace(10, 35, 10))
```

## Multi-facet grids

Use two facet quantities to make a grid of faceted plots

```python
Facet([data.cyl, data.am], [data.wt, data.mpg]).scatter()
```

## More control

Some plots need more than a single function call. You can iterate over a facet object, to customize your plots

```python
f = Facet(data.cyl, [data.wt, data.mpg])
for item in f:
    #like the facet object, this auto-applies the data arguments
    item.scatter()

    axes = item.axes
    #tuple of the subset of the data for this facet
    x, y = item.data
    
    m, b = np.polyfit(x, y, 1)
    yp = m * x + b    
    axes.plot(x, yp, 'r', lw=3)

    print "Facet value is %s. Label is %s" % (item.key, item.label)
```

## Labeling facets

```python
labeler = {4:'Four cylinder', 6:'Six cylinder', 8:'Eight cylinder'}
f = Facet(data.cyl, [data.wt, data.mpg], labeler=labeler).plot('o')
```

Labelers can also be functions

```python
def labeler(key):
    k1, k2 = key
    am = 'Automatic' if k2 == 0 else 'Manual'
    return "Cyl:%s (%s)" % (k1, am)

f = Facet([data.cyl, data.am], [data.mpg], labeler=labeler, xlabel='MPG')
f.hist(bins=np.linspace(10, 30, 8))
```

## Working with labled arrays

For objects with labeled columns (i.e. pandas DataFrames, astropy tables, numpy recarrays), the `from_labels` class method is helpful:

### Pandas

```python
import pandas as pd
df = pd.DataFrame(data)
f = Facet.from_labels(df, ['cyl', 'am'], ['mpg'])
f.hist(bins=np.linspace(10, 30, 8))
```

### Astropy tables

```python
from astropy.table import Table
tbl = Table(data)
f = Facet.from_labels(tbl, ['cyl', 'am'], ['mpg'])
f.hist(bins=np.linspace(10, 30, 8))
```

### Numpy recarrays

```python
f = Facet.from_labels(data, ['cyl', 'am'], ['mpg'])
f.hist(bins=np.linspace(10, 30, 8))
```
